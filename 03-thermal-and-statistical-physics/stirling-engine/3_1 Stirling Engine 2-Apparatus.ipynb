### Piston and crank kinematics

#### Piston position

Suppose the crank rod has length $L$ and is attached to the crank at radius $R$ from the crank axis. Then the height $z$ of the piston above the crank axis is related to $L$ and $R$ by the law of cosines:

$$ L^{2}=R^{2}+z^{2}-2Rz\cos \theta .$$

Solve for z (this can also be written down directly from a geometric perspective):

$$ z=R\cos\theta+\sqrt{L^{2}-R^{2}\sin^{2}\theta} . $$

$$ z=R\cos\theta+L\sqrt{1-\left (\frac{R}{L}  \right )^{2}\sin^{2}\theta}. $$

For $L>>R$

$$ z\approx L+R\cos\theta .$$

#### Volume

We assume that when the crank is "top dead center" (maximum z), the working gas volume within the cylinder is compressed to a minimum value $ V_\mathrm{min} $ that was designed into the system. Then as the piston moves away from its top dead center position, the volume increases to a value $V$ given by:

$$ V=V_\mathrm{min}+\left ( L+R-z \right )A .$$

Notice that the maximum working gas volume $V_\mathrm{max}$ is obtained when $\theta = \pi $ so that $ z = L - R $:

$$ V_\mathrm{max}=V_\mathrm{min}+2RA .$$

#### Angle relations

$$ \sin \phi =\frac{R\sin\theta}{L}=\frac{R}{L}\sin\theta . $$

$$ \cos \phi =\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin ^{2}\theta} .$$

$$ \alpha= \pi-(\theta+\phi).$$

$$ \beta= \pi-\alpha=\theta+\phi .$$


#### Time derivatives

With some steps omitted:

$$\dot{z}=-\sin\theta\left [ 1+\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}} \right ]R\dot{\theta} .$$

$$ \ddot{z}=-\left \{ \cos\theta+\frac{\frac{R}{L}\left ( 1-2\sin^{2}\theta \right )+\left ( \frac{R}{L} \right )^{3}\sin^{4}\theta}{\left [ 1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta \right ]^{3/2}} \right \}R\dot{\theta}^{2}-\sin\theta\left [ 1+\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}} \right ]R\ddot{\theta} . $$

Notice that when $\theta=0$ and $\theta=\pi$ then

$$ \ddot{z}=\left(\mp 1-\frac{R}{L}\right)R\dot{\theta}^{2} $$

respectively.


### Piston Dynamics

The system diagram shows three forces acting on the piston: the connecting rod force $F_{\mathrm{rod}}$, the net pressure force driving the piston $F_{\mathrm{piston}}$, and the friction force $ f_{\mathrm{piston}} $ due to the seals sliding against the cylinder wall.

Newton's Second Law applied to the piston determines the piston acceleration $ \ddot{z} $:

$$ F_{\mathrm{rod}}\cos\phi-F_{\mathrm{piston}}\pm f_{\mathrm{piston}}=m_{\mathrm{piston}}\ddot{z} . $$

Here the piston force $F_{\mathrm{piston}}$ is the difference between the cylinder pressure $p$ and lab pressure $p_{\mathrm{lab}}$ multiplied by the piston area $A$:

$$ F_{\mathrm{piston}}=\left( p - p_{\mathrm{lab}} \right) A ,$$

In the equation for Newton's second law, two choices of sign are shown for the friction force $ f_{\mathrm{piston}} $: the top choice $+$ applies to the friction force $f_{\mathrm{piston}}$ when the piston is traveling downward to decrease piston height $z$ (power stroke) and the bottom choice $-$ applies to the friction force $f_{\mathrm{piston}}$ when the piston is traveling upward to increase piston height $z$ (return stroke).

For use below, we'll solve this now for $F_{\mathrm{rod}}$:

$$F_{\mathrm{rod}}=\frac{1}{\cos\phi}\left[\left( p - p_{\mathrm{lab}} \right) A \mp f_{\mathrm{piston}}+m_{\mathrm{piston}}\ddot{z}\right]$$

### Flywheel Dynamics

The connecting rod applies a torque $/tau$ to the crank. If the crank is attached to a flywheel so that the rotating system has moment of inertia $I$ with bearing frictional torque $\tau_{\mathrm{friction}}$ and a load torque $\tau_{\mathrm{load}}$ (e.g. due to a cord lifting up a weight)  then

$$\tau-\tau_{\mathrm{friction}}-\tau_{\mathrm{load}}=I\ddot{\theta}$$
The torque on the flywheel is given by

$$ \tau=F_{\mathrm{rod}}R\sin\beta . $$

$$ \tau=F_{\mathrm{rod}}R\sin(\theta+\phi) .$$

$$ \tau=F_{\mathrm{rod}}R\left (\sin\theta\cos\phi+\cos\theta\sin\phi  \right ) .$$

$$ \tau=\frac{1}{\cos\phi}\left[\left( p - p_{\mathrm{lab}} \right) A \mp f_{\mathrm{piston}}+m_{\mathrm{piston}}\ddot{z}\right]R(\sin\theta\cos\phi+\cos\theta\sin\phi) . $$

$$ \tau=\left[\left( p - p_{\mathrm{lab}} \right) A \mp f_{\mathrm{piston}}+m_{\mathrm{piston}}\ddot{z}\right]R\left(\sin\theta+\cos\theta\frac{\sin\phi}{\cos\phi}\right) . $$

$$ \tau=\left[\left( p - p_{\mathrm{lab}} \right) A \mp f_{\mathrm{piston}}+m_{\mathrm{piston}}\ddot{z}\right]R\left [ \sin\theta +\cos\theta\frac{\frac{R}{L}\sin\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ] .$$

Thus the flywheel dynamics becomes

$$\left[\left( p - p_{\mathrm{lab}} \right) A \mp f_{\mathrm{piston}}+m_{\mathrm{piston}}\ddot{z}\right]R\sin\theta\left [ 1 +\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ]-\tau_{\mathrm{friction}}-\tau_{\mathrm{load}}=I\ddot{\theta} .$$

### Relating pressure to piston position

$$p=\frac{n\mathbb{R}T}{V}.$$

$$p=\frac{n\mathbb{R}T}{V_\mathrm{min}+\left ( L+R-z \right )A}.$$

We assume the following:

$$ p_{\mathrm{lab}}=\frac{n\mathbb{R}T_{\mathrm{cool}}}{V_\mathrm{max}}. $$

So

$$ p_{\mathrm{lab}}=\frac{n\mathbb{R}T_{\mathrm{cool}}}{V_{\mathrm{min}}+2RA}$$

and then

$$ V_{\mathrm{min}}=\frac{n\mathbb{R}T_{\mathrm{cool}}}{p_{\mathrm{lab}}}-2RA.$$

Thus we can write an expression for the general pressure $p$ as

$$ p=\frac{n\mathbb{R}T}{\frac{n\mathbb{R}T_{\mathrm{cool}}}{p_{\mathrm{lab}}}+\left ( L-R-z \right )A}.$$

and then

$$p=\frac{T}{T_{\mathrm{cool}}+\frac{p_{\mathrm{lab}}AL}{n\mathbb{R}}\left ( 1-\frac{R}{L}-\frac{z}{L} \right )}p_{\mathrm{lab}}.$$

### New dynamical equation

Now use

$$ F_{\mathrm{piston}}=\left[ \frac{T}{T_{\mathrm{cool}}+\frac{p_{\mathrm{lab}}AL}{n\mathbb{R}}\left ( 1-\frac{R}{L}-\frac{z}{L} \right )} - 1 \right]p_{\mathrm{lab}} A $$

to get

$$\left\{\left[ \frac{T}{T_{\mathrm{cool}}+\frac{p_{\mathrm{lab}}AL}{n\mathbb{R}}\left ( 1-\frac{R}{L}-\frac{z}{L} \right )} - 1 \right]p_{\mathrm{lab}} A \mp f_{\mathrm{piston}}+m_{\mathrm{piston}}\ddot{z}\right\}R\sin\theta\left [ 1 +\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ]-\tau_{\mathrm{friction}}-\tau_{\mathrm{load}}=I\ddot{\theta} .$$

Put all the time-derivatives on the left:

$$I\ddot{\theta}-m_{\mathrm{piston}}\ddot{z}R\sin\theta\left [ 1 +\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ]=\\ \left\{\left[ \frac{T}{T_{\mathrm{cool}}+\frac{p_{\mathrm{lab}}AL}{n\mathbb{R}}\left ( 1-\frac{R}{L}-\frac{z}{L} \right )} - 1 \right]p_{\mathrm{lab}} A \mp f_{\mathrm{piston}}\right\}R\sin\theta\left [ 1 +\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ]-\tau_{\mathrm{friction}}-\tau_{\mathrm{load}} .$$

Replace all z-dependent terms with the angular equivalents

$$ I\ddot{\theta}+m_{\mathrm{piston}}R^{2}\sin\theta \left [ 1 +\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ]\left \{ \cos\theta+\frac{\frac{R}{L}\left ( 1-2\sin^{2}\theta \right )+\left ( \frac{R}{L} \right )^{3}\sin^{4}\theta}{\left [ 1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta \right ]^{3/2}} \right \}\dot{\theta}^{2}\\+m_{\mathrm{piston}}R^{2}\sin^{2}\theta\left [ 1+\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}} \right ]^{2}\ddot{\theta}=\left\{\left[ \frac{T}{T_{\mathrm{cool}}+\frac{p_{\mathrm{lab}}AL}{n\mathbb{R}}\left (1-\left(\frac{R}{L} \right )-\left(\frac{R}{L} \right )\cos\theta-\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}\right )} - 1 \right]p_{\mathrm{lab}} A \mp f_{\mathrm{piston}}\right\}\\ \times R\sin\theta\left [ 1 +\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ]-\tau_{\mathrm{friction}}-\tau_{\mathrm{load}} $$

And finally:

$$\left\{I+m_{\mathrm{piston}}R^{2}\sin^{2}\theta\left [ 1+\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}} \right ]^{2}\right\}\ddot{\theta}\\+m_{\mathrm{piston}}R^{2}\sin\theta \left [ 1 +\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ] \left \{ \cos\theta+\frac{\frac{R}{L}\left ( 1-2\sin^{2}\theta \right )+\left ( \frac{R}{L} \right )^{3}\sin^{4}\theta}{\left [ 1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta \right ]^{3/2}} \right \}\dot{\theta}^{2}=\left\{\left[ \frac{T}{T_{\mathrm{cool}}+\frac{p_{\mathrm{lab}}AL}{n\mathbb{R}}\left (1-\left(\frac{R}{L} \right )\left(1+\cos\theta\right)-\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}\right )} - 1 \right]p_{\mathrm{lab}} A \mp f_{\mathrm{piston}}\right\}\\ \times R\sin\theta\left [ 1 +\frac{\frac{R}{L}\cos\theta}{\sqrt{1-\left ( \frac{R}{L} \right )^{2}\sin^{2}\theta}}\right ]-\tau_{\mathrm{friction}}-\tau_{\mathrm{load}} .$$

If we keep lowest order terms for $R << L$, this simplifies to:

$$ \left(I+m_{\mathrm{piston}}R^{2}\sin^{2}\theta\right)\ddot{\theta}+m_{\mathrm{piston}}R^{2}\sin\theta   \cos\theta \dot{\theta}^{2}\\=\left\{\left[ \frac{T}{T_{\mathrm{cool}}-\frac{p_{\mathrm{lab}}AR(1+\cos\theta)}{n\mathbb{R}}} - 1 \right]p_{\mathrm{lab}} A \mp f_{\mathrm{piston}}\right\}  R\sin\theta-\tau_{\mathrm{friction}}-\tau_{\mathrm{load}} $$
